```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.0, 2.0, 0.01)
y = 1 + np.sin(2*np.pi*x)

plt.plot(x, y)
plt.title('Sinus chart')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
